```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np

df = pd.read_csv("Data/Auto.csv")
df.head()
```

```python
df = df.replace("?", np.nan)
df["horsepower"] = df["horsepower"].fillna(df.mean())
```

```python
quantitative = df.copy().drop(["name", "origin"], axis=1)
quantitative.max() - quantitative.min()
```

```python
quantitative.describe()
```

```python
quantitative.drop(index=list(range(10, 85))).describe()
```
